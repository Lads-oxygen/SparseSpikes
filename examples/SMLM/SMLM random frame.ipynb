```julia
using Revise, Plots, Images
```

Define ground truth amplitudes and positions

```julia
# Define the plot
num_points = 64

domain = [[0, 1], [0, 1]]

plt_grid_x1 = [domain[1][1] + i * (domain[1][2] - domain[1][1]) / (num_points - 1) for j in 0:num_points-1, i in 0:num_points-1]
plt_grid_x2 = [domain[2][1] + j * (domain[2][2] - domain[2][1]) / (num_points - 1) for j in 0:num_points-1, i in 0:num_points-1]

grid = range(0, stop=1, length=num_points)
plot_size = (400, 400) .* 2
plt = heatmap(xlims=domain[1], ylims=domain[2], xlabel="", ylabel="", ticks=false, color=:viridis, colorbar=false, size=plot_size, grid=false, ratio=:equal)

# plt_gt = deepcopy(plt)
# plot_spikes!(plt_gt, μ0, label=L"μ_0", colorscheme=:viridis)
```

```julia
image = load("../sequence/01386.tif")


y = channelview(image)

using Plots.Measures
heatmap!(plt, grid, grid, y, label="y_0", color=:viridis, margins=-2mm)
```
